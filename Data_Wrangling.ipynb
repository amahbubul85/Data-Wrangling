# Import Pandas and Matplotlib.pyplot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Path for Automobile Dataset

```python
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

# Python list for headers

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

# Use the Pandas method read_csv() to load the data from the web address and set the column names

```python
df=pd.read_csv(filename,names=headers)
```

# Checking for missing values

```python
list_columns=df.columns.values.tolist()
for item in list_columns:
    print(df[item].value_counts())
```

# Importing Numpy library

```python
import numpy as np
```

# Replace the Question marks by NaN

```python
for item in list_columns:
    df[item].replace('?',np.nan,inplace=True)
```

# Checking the Data for NaN and counting them

```python
Missing_values=df.isnull()
for item in Missing_values.columns:
    print(Missing_values[item].value_counts())
```

# Replacing NaN in 'normalized-losses' by mean of the column

```python
mean_normalized_losses=df['normalized-losses'].astype(float).mean()
mean_stroke=df["stroke"].astype(float).mean()
mean_bore=df["bore"].astype(float).mean()
mean_horsepower=df["horsepower"].astype(float).mean()
mean_peak_rpm=df["peak-rpm"].astype(float).mean()
print(mean_normalized_losses)
print(mean_stroke)
print(mean_bore)
print(mean_horsepower)
print(mean_peak_rpm)

df['normalized-losses'].replace(np.nan,mean_normalized_losses,inplace=True)
df["stroke"].replace(np.nan,mean_stroke,inplace=True)
df["bore"].replace(np.nan,mean_bore,inplace=True)
df["horsepower"].replace(np.nan,mean_horsepower,inplace=True)
df["peak-rpm"].replace(np.nan,mean_peak_rpm,inplace=True)

print(df['normalized-losses'].value_counts())
print(df["stroke"].value_counts())
print(df["bore"].value_counts())
print(df["horsepower"].value_counts())
print(df["peak-rpm"].value_counts())
```

# replace the missing 'num-of-doors' values by the most frequent

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

# replace the rows having any NaN for the column 'price'

```python
df.dropna(axis=0, how='any', thresh=None, subset=['price'], inplace=True)
```

# reset index, because we droped two rows

```python
df.reset_index(drop=True, inplace=True)
```

# Changing the type to the correct one

```python
df[['peak-rpm','symboling']] = df[['peak-rpm','symboling']].astype("int")
df.dtypes
```

# Data Standardization: Data transformation


```python
df['city-L/100km'] = 235/df["city-mpg"]
df['highway-mpg/100km'] = 235/df["highway-mpg"]
df.dtypes
```

# Data Standardization: Data normalization

```python
df['height']/df['height'].max()
df['height']/df['height'].max()
df['height']/df['height'].max()
```

# Data Standardization: Data binning

```python
plt.hist(df['city-mpg'],bins=3)
bins=np.linspace(df['city-mpg'].min(),df['city-mpg'].max(),4)
group_names=['low','medium','high']
df['city-mpg_bins']=pd.cut(df['city-mpg'],bins,labels=group_names,include_lowest=True)
df.head()
```

```python
plt.bar(group_names,df['city-mpg_bins'].value_counts())
```

# Indicator variable (or dummy variable)

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()

# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```
